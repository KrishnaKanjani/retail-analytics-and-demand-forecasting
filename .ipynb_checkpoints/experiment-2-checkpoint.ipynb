```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataset.csv", encoding="ISO-8859-1")
data.head(5)
```

```python
# Convert order date to datetime format
data['order date (DateOrders)'] = pd.to_datetime(data['order date (DateOrders)'])

# Aggregate data by day and sum sales
daily_data = data.groupby(pd.Grouper(key='order date (DateOrders)', freq='D')).agg({'Sales': 'sum'}).reset_index()
```

```python
daily_data
```

```python
daily_data = daily_data[daily_data['order date (DateOrders)'] < '2018-01-01']
daily_data.rename(columns={'order date (DateOrders)': 'ds', 'Sales': 'y'}, inplace=True)
daily_data.tail(30)
```

```python
daily_data['YearMonth'] = daily_data['ds'].dt.to_period('M')

# Aggregate data by year-month and sum sales
monthly_sales = daily_data.groupby('YearMonth')['y'].sum()
```

```python
plt.figure(figsize=(15, 6))
monthly_sales.plot(kind='line', color='b')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(monthly_sales.index, [x.strftime('%b %Y') for x in monthly_sales.index], rotation=90)
plt.show()
```

```python
from prophet import Prophet
```

```python
holidays_df = pd.read_excel('US-PR-holidays.xlsx')
holidays_df = holidays_df.rename(columns={'Date': 'ds',
                                          'Name': 'holiday'})
holidays_df['ds'] = pd.to_datetime(holidays_df['ds'])
holidays_df = holidays_df.drop(columns=['Country', 'Year'])

# Add 'lower_window' and 'upper_window' columns
holidays_df['lower_window'] = 0
holidays_df['upper_window'] = 1

model = Prophet(holidays=holidays_df,
                holidays_prior_scale=15)
model.fit(daily_data)
```

```python
future = model.make_future_dataframe(periods=365)
```

```python
forecast = model.predict(future)
```

```python
forecast[['ds', 'yhat']].head(10)
# fig = model.plot(forecast)
```

```python
daily_data.head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_data['ds'], daily_data['y'], label='Actual Sales', color='blue')
plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='red')
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='pink', alpha=0.3)
plt.title('Daily Sales Data vs. Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
daily_data['YearMonth'] = daily_data['ds'].dt.to_period('M')
actual = daily_data.groupby('YearMonth')['y'].sum().reset_index()
actual['YearMonth'] = actual['YearMonth'].astype(str)

forecast['YearMonth'] = forecast['ds'].dt.to_period('M')
monthly_sales = forecast.groupby('YearMonth')['yhat'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str)

plt.figure(figsize=(18, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['yhat'], linestyle='-', color='red', label='Forecast') #forecasted
plt.plot(actual['YearMonth'], actual['y'], linestyle='-', color='green', label='Actual')  #actual
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
from prophet.diagnostics import cross_validation, performance_metrics
cv = cross_validation(model=model, initial='365 days', period='365 days', horizon='365 days')
```

```python
pm = performance_metrics(cv, rolling_window=0)
```

```python
pm['mape'].mean()
```

```python
from prophet.plot import plot_cross_validation_metric

plot_cross_validation_metric(cv, metric='mape');
```
