```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load your dataset
dataset = pd.read_csv("dataset.csv", encoding="ISO-8859-1")
dataset.head(5)
```

```python
dataset["order date (DateOrders)"] = pd.to_datetime(dataset["order date (DateOrders)"])
dataset['Order-Year'] = dataset['order date (DateOrders)'].dt.year
dataset['Order-Month'] = dataset['order date (DateOrders)'].dt.month
dataset['Order-day'] = dataset['order date (DateOrders)'].dt.day
dataset['Order-weekday'] = dataset['order date (DateOrders)'].dt.weekday
```

```python
# Assuming 'product_id' identifies the top-selling product
top_product_data = dataset[dataset['Product Card Id'] == 365]

# Exclude years 2017 and 2018 from the data
#top_product_data = top_product_data[(top_product_data['Order-Year'] >= 2015) & (top_product_data['Order-Year'] <= 2017)]
top_product_data
```

```python
top_product_data['order date (DateOrders)'] = pd.to_datetime(top_product_data['order date (DateOrders)'])
```

```python
# Explore data for patterns, seasonality, and trends (e.g., using plots)
plt.figure(figsize=(10, 6))
df_lineplot = top_product_data[['order date (DateOrders)', 'Sales']]

# Set 'order date (DateOrders)' as index
df_lineplot.set_index('order date (DateOrders)', inplace=True)

# Resample to monthly frequency and sum the sales
monthly_sales = df_lineplot.resample('M').sum()

# Plot the time series line plot
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Sales Trend of Top-Selling Product')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)

# Format the x-axis tick labels to display month and year
plt.xticks(monthly_sales.index, [x.strftime('%b %Y') for x in monthly_sales.index], rotation=90)
plt.show()
```

```python
# Based on the insights, select a suitable forecasting model (e.g., Prophet)
from prophet import Prophet
```

```python
# Prepare data in required format for Prophet
prophet_data = top_product_data.copy()
prophet_data = prophet_data.rename(columns={'order date (DateOrders)': 'ds', 'Sales': 'y'})

holidays_df = pd.read_excel('US-PR-holidays.xlsx')
holidays_df = holidays_df.rename(columns={'Date': 'ds',
                                          'Name': 'holiday'})
holidays_df['ds'] = pd.to_datetime(holidays_df['ds'])
holidays_df = holidays_df.drop(columns=['Country', 'Year'])

# Add 'lower_window' and 'upper_window' columns
holidays_df['lower_window'] = 0
holidays_df['upper_window'] = 1
```

```python
# Initialize and fit Prophet model for forecasting
model = Prophet(seasonality_mode= 'additive',
                seasonality_prior_scale= 5,
                holidays_prior_scale= 1,
                changepoint_prior_scale= 0.1)

model.add_country_holidays(country_name='PR')
model.add_country_holidays(country_name='US')

model.fit(prophet_data)
```

```python
from prophet.diagnostics import cross_validation, performance_metrics

future_dates = model.make_future_dataframe(periods=730, freq='D')
forecast = model.predict(future_dates)
```

```python
cv = cross_validation(model=model, initial='365 days', period='365 days', horizon='365 days')
```

```python
pm = performance_metrics(cv, rolling_window=0)
```

```python
from prophet.plot import plot_cross_validation_metric

plot_cross_validation_metric(cv, metric='mape');
```

```python
pm['mape'].mean()
```

```python
plt.figure(figsize=(18, 6))
# Plot the forecast before 2018 in blue
plt.plot(forecast[forecast['ds'] < '2018-01-01']['ds'], forecast[forecast['ds'] < '2018-01-01']['yhat'], linestyle='-', color='blue')
# Plot the forecast after 2018 in red
plt.plot(forecast[forecast['ds'] >= '2018-01-01']['ds'], forecast[forecast['ds'] >= '2018-01-01']['yhat'], linestyle='-', color='green')

# plt.xlabel('Date')
# plt.ylabel('Sales')
plt.title('Daily Sales Forecast for Next 24 Months')
plt.grid(True)
# plt.legend()
plt.show()
```

```python
forecast['year_month'] = forecast['ds'].dt.to_period('M').astype(str)
forecast_monthly_sales = forecast.groupby('year_month')['yhat'].sum().reset_index()

top_product_data['year_month'] = top_product_data['order date (DateOrders)'].dt.to_period('M').astype(str)
actual_monthly_sales = top_product_data.groupby('year_month')['Sales'].sum().reset_index()

plt.figure(figsize=(15, 5))
plt.plot(forecast_monthly_sales['year_month'], forecast_monthly_sales['yhat'], linestyle='-', color='green', label='forecast monthly sales')
plt.plot(actual_monthly_sales['year_month'], actual_monthly_sales['Sales'], linestyle='-', color='blue', label='actual monthly sales')  #actual
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
